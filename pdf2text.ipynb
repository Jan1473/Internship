```python
import re
from pdfminer.high_level import extract_pages,extract_text
```

```python
for page_layout in extract_pages("qualityhosting_test.pdf"):
    for element in page_layout:
        print(element)
```

```python
text=extract_text("qualityhosting_test.pdf")
print(text)
```

```python
import re
invoice_no=re.compile(r'^\d{8}')
INVOICE_NO=0
```

```python
for line in text.split('\n'):
    if invoice_no.match(line):
        INVOICE_NO=line
```

```python
date=re.compile(r'\d{2}.\d{2}.\d{2}')
Date=0
```

```python
for line1 in text.split('\n'):
    if date.match(line1):
        Date=line1
        
```

```python
total_amt=re.compile(r'Total\sEUR')
TOTAL_AMT=[]
```

```python
for line2 in text.split('\n'):
    if total_amt.match(line2):
        print(line2)
        TOTAL_AMT=line2
```

```python
tot1=re.compile(r'\d{2},\d{2}')
tot=0
for line3 in text.split('\n'):
    if tot1.match(line3):
        print(line3)
        tot=line3
```
